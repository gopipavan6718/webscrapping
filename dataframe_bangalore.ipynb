```python
import requests
from bs4 import BeautifulSoup
import re
import pandas as pd
import numpy as np
```

```python
Bedrooms = []
Prices = []
Locality = []
Sqft = []
Status = []
Persqft =[]
lst = ['https://www.makaan.com/bangalore-residential-property/buy-property-in-bangalore-city?propertyType=villa',
      'https://www.makaan.com/bangalore-residential-property/buy-property-in-bangalore-city?propertyType=villa&page=3',
      'https://www.makaan.com/bangalore-residential-property/buy-property-in-bangalore-city?propertyType=villa&page=4',
      'https://www.makaan.com/bangalore-residential-property/buy-property-in-bangalore-city?propertyType=villa&page=5',
      'https://www.makaan.com/bangalore-residential-property/buy-property-in-bangalore-city?propertyType=villa&page=6',
      'https://www.makaan.com/bangalore-residential-property/buy-property-in-bangalore-city?propertyType=villa&page=7',
      'https://www.makaan.com/bangalore-residential-property/buy-property-in-bangalore-city?propertyType=villa&page=8',
      'https://www.makaan.com/bangalore-residential-property/buy-property-in-bangalore-city?propertyType=villa&page=9',
      'https://www.makaan.com/bangalore-residential-property/buy-property-in-bangalore-city?propertyType=villa&page=10',
      'https://www.makaan.com/bangalore-residential-property/buy-property-in-bangalore-city?propertyType=villa&page=11',
      'https://www.makaan.com/bangalore-residential-property/buy-property-in-bangalore-city?propertyType=villa&page=12',
      'https://www.makaan.com/bangalore-residential-property/buy-property-in-bangalore-city?propertyType=villa&page=13',
      'https://www.makaan.com/bangalore-residential-property/buy-property-in-bangalore-city?propertyType=villa&page=14']
for i in lst:
    print(i)
    url = i
    page = requests.get(url)
    page.text
    soup = BeautifulSoup(page.text)


    p = soup.find_all("div",class_="title-line")


    for i in p:
        a = i.text
        b = re.findall(r'(\d+\s+BHK)',a)
        Bedrooms.append(b[0])

    p = soup.find_all("td",class_="price")


    for i in p:
        a = i.text

        Prices.append(a)



    p = soup.find_all("span",class_="locName")

    for i in p:
        a = i.text
        Locality.append(a)


    p = soup.find_all("td",class_="size")


    for i in p:
        a = i.text
        Sqft.append(a)


    p = soup.find_all("td",class_="val")


    for i in p:
        a = i.text
        Status.append(a)


    p = soup.find_all("td",class_="lbl rate")


    for i in p:
        a = i.text
        Persqft.append(a)
    Persqft



```

```python
df3 = pd.DataFrame({'Bedrooms':Bedrooms,
                  
                  'Locality':Locality,
                  'Sqft':Sqft,
                  'Status':Status,
                  'Persqft':Persqft,
                  'Prices':Prices
                   
})

df3
```

```python
df3[['Location','City']]=df3['Locality'].str.split(',',expand = True)
df3
```

```python
df3.drop(columns = 'Locality',inplace = True)
df3
```

```python
df = pd.read_csv('result_csv')
df.drop(columns = ['Unnamed: 0'],inplace = True)
```

```python
final = pd.concat([df,df3],axis = 0)
```

```python
final=final.reset_index(drop = True)
final
```
